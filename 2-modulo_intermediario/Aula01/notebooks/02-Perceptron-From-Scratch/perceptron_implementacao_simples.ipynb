## Perceptron implementaao em Python

```python
dimension = 2
w = [2*random() - 1 for i in range(dimension)]
b = 2*random() - 1
```

```python
learning_rate = 1.0
epochs = 100
for step in range(epochs):
    cost = 0
    for x_n, y_n in zip(x, y):
        y_pred = sum([x_i * w_i for x_i, w_i in zip(x_n, w)]) + b
        y_pred = 1 if y_pred > 0 else 0
        error = y_n - y_pred
        w = [w_i + learning_rate * error * x_i for x_i, w_i in zip(x_n, w)]
        b = b + learning_rate * error
        cost += error**2
        
    if step % 10 == 0:
        print('step {0}: {1}'.format(step, cost))
          
preds = [1 if y > 0 else 0 for y in np.dot(x, np.array(w)) + b]

print('w: ', w)
print('b:', b)
print('y_pred: {0}'.format(preds))
```

```python
matrix = confusion_matrix(preds,y, labels=[1,0])
print('Confusion matrix : n',matrix)
```

```python
#Plotting the confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(matrix, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

## Perceptron implementao usando Numpy

```python
dimension = x.shape[1]
w = 2*np.random.random(size=dimension)-1
b = 2*np.random.random()-1 

learning_rate = 1.0
epochs = 10
for step in range(epochs):
    cost = 0
    for x_n, y_n in zip(x, y):
        y_pred = np.dot(x_n, w) + b 
        y_pred = np.where(y_pred > 0, 1, 0)
        error = y_n - y_pred
        w = w + learning_rate * np.dot(error, x_n)
        b = b + learning_rate * error
        cost += error**2
    
    if step%10 == 0:
        print('step {0}: {1}'.format(step, cost))
    
preds = [1 if y > 0 else 0 for y in np.dot(x, np.array(w)) + b]

print('w: ', w)
print('b:', b)
print('y_pred: {0}'.format(preds))
```

```python
matrix = confusion_matrix(preds,y, labels=[1,0])
print('Confusion matrix : n',matrix)
```

```python
#Plotting the confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(matrix, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```
